```python
##visulizing data
##Comparing dimension with species
sns.set_style('whitegrid')
sns.lmplot(y='SepalWidthCm',x='SepalLengthCm',data=df,fit_reg=False,hue='Species')
plt.title("Scatter Plot: Sepal Width vs Sepal Lenght(cm)")
```

```python
sns.set_style('whitegrid')
sns.lmplot(y='PetalWidthCm',x='PetalLengthCm',data=df,fit_reg=False,hue='Species')
plt.title("Scatter Plot: Petal Width vs Petal Lenght(cm)")
plt.legend()
```

```python
from sklearn import preprocessing 
X=df.iloc[:,:-1].values
css = []
for i in range(1,15): 
    #Building and fitting the model 
    kmeanModel = KMeans(n_clusters=i)
    kmeanModel.fit(X) 
    css.append(kmeanModel.inertia_)
print(css)
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(10,5))
plt.plot(range(1,15),css)
plt.title('The elbow method',size=15)
plt.xlabel('Number of clusters',size=12)
plt.ylabel('WCSS',size=12) # Within cluster sum of squares
plt.show()
```

```python
kmeans=KMeans(n_clusters=3)
classifier=kmeans.fit_predict(X)

df=df.copy()
df['Cluster']=classifier
df.head()
```

```python
#cluster visulization
sns.set_style('whitegrid')
plt.figure(figsize=(7,6))
plt.title('Clusters w.r.t. features Sepal length and Petal length \n')

plt.scatter(X[classifier==0,0],X[classifier==0,2],s=70,c='green',label='Iris-Setosa')
plt.scatter(X[classifier==1,0],X[classifier==1,2],s=70,c='yellow',label='Iris-Virginica')
plt.scatter(X[classifier==2,0],X[classifier==2,2],s=70,c='purple',label='Iris-Versicolor')

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,2],s=100,c='black',label='Centroids')

plt.legend()
plt.show()
```
