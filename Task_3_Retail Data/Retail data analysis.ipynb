```python
df.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['green','purple'],figsize=(10,8))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

```python
plt.figure(figsize=(8,9))
plt.title('Segment wise Sales in each Region')
sns.barplot(x='Sales',y='Region',data=df,hue='Segment',order=df['Region'].value_counts().index,palette='rocket')
plt.xlabel('Region',fontsize=5)
plt.show()
```
